```python
import pandas as pd
import numpy as np

houses = pd.read_csv('housing_data.csv')
```

```python
# Do what we did last week with numpy
# Create a matrix (numpy array object type) from our houses
houses.reindex(np.random.permutation(houses.index))
np_feature = np.array(houses['sqft_living'])
np_target = np.array(houses['price'])

pivot = int(len(np_feature) * 0.8)
np_feature, np_test_feature = np_feature[:pivot], np_feature[pivot:]
np_target, np_test_target = np_target[:pivot], np_target[pivot:]

print np_feature[:3], np_target[:3]
print np_feature.shape, np_target.shape
```

```python
# We do this because numpy and sklearn both assume multiple features
A = np.vstack([np_feature, np.ones(len(np_feature))]).T
```

```python
m, c = np.linalg.lstsq(A, np_target)[0]
print m, c
```

```python
# Lets visualize the results
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(np_feature, np_target, 'o', alpha=0.3)
plt.plot(np_feature, m*np_feature + c, 'r')
```

```python
# Now lets do it was SciKit Learn
from sklearn import linear_model

model = linear_model.LinearRegression()
model.fit(A, np_target)
```

```python
# make a prediction
expected = np_test_target
A_test = np.vstack([np_test_feature, np.ones(len(np_test_feature))]).T

predicted = model.predict(A_test)

# summarize how well our model does
mss = np.mean((predicted - expected)**2)
print mss
print "Our model is correct %s percent of the time" % model.score(A_test, np_test_target)
```

```python
# Lets try multiple features with SciKit Learn
# we want to create some new features, the target (price) will be the same
houses.reindex(np.random.permutation(houses.index))
feature_labels = ['sqft_living', 'sqft_lot', 'sqft_basement', 'sqft_above', 'bedrooms', 'long', 'zipcode']

pd_features = houses[feature_labels]
features = np.array(pd_features)
targets = np.array(houses['price'])

pivot = len(features) * .8
train_features, test_features = features[:pivot], features[pivot:]
train_target, test_target = targets[:pivot], targets[pivot:]
```

```python
# Train the model with multiple features
model = linear_model.LinearRegression()
model.fit(train_features, train_target)
```

```python
# predict values
expected = test_target
predicted = model.predict(test_features)

# summarize the data
mss = np.mean((predicted - expected)**2)
print mss
print model.score(test_features, test_target)
```
